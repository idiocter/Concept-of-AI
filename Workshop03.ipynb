##Which passenger had the highest fare paid relative to their age?
To answer the question perform following operations:
1. Add a column to the Titanic dataset, fare per year, containing the fare divided by the age of
the passenger(i.e., Fare/Age).
2. Subset rows where fare per year is higher than 5, assigning this to high fare age.
3. Sort high fare age by descending fare per year, assigning this to high fare age srt.
4. Select only the Name and fare per year columns of high fare age srt and save the result as result.
5. Look at the result.

```
from IPython.core.interactiveshell import dis
"""
Problem: Identify the passenger who paid the highest fare relative to their age.

Steps performed:
1. Create a 'FarePerYear' column by dividing 'Fare' by 'Age'.
2. Subset passengers where 'FarePerYear' > 5.
3. Sort this subset in descending order of 'FarePerYear'.
4. Select only 'Name' and 'FarePerYear' columns to display the result.

Author: Bipul Regmi
"""

#Fill missing Age values with median to avoid division by NaN
titanic_df["Age"] = titanic_df["Age"].fillna(titanic_df["Age"].median())

#Add column 'FarePerYear'
titanic_df["FarePerYear"] = titanic_df["Fare"] / titanic_df["Age"]
print("\n\nFarePerYear column:")
display(titanic_df["FarePerYear"])

#Subset rows where FarePerYear > 5
high_fare_age = titanic_df[titanic_df["FarePerYear"] > 5]
print("\n\nPassengers with FarePerYear > 5:")
display(high_fare_age)

#Sort by FarePerYear descending
high_fare_age_srt = high_fare_age.sort_values(by="FarePerYear", ascending=False)
print("\n\nPassengers with FarePerYear > 5 sorted by FarePerYear:")
display(high_fare_age_srt)

#Select only 'Name' and 'FarePerYear'
result = high_fare_age_srt[["Name", "FarePerYear"]]

#View the result
print("\n\nPassengers with highest fare relative to age:")
display(result)
```

##Which adult male passenger (age  18 and Sex is male) paid the highest fare relative to their class?
To answer the question perform following operations:
1. Add a column to the Titanic dataset, fare per class, containing the fare divided by the passenger
class i.e. Fare / Pclass.
2. Subset rows where the passenger is male (Sex is male) and an adult (Age is greater than or equal
to 18), assigning this to adult males.
3. Sort adult males by descending fare per class, assigning this to adult males srt.
4. Select only the Name, Age, and fare per class columns of adult males sr and save the result
as result.

```
"""
Problem: Identify the adult male passenger who paid the highest fare relative to their class.

Workflow:
1. Create a new column 'FarePerClass' as Fare divided by Pclass.
2. Filter the dataset to include only adult male passengers (Age  18 and Sex = 'male').
3. Sort the filtered dataset in descending order of 'FarePerClass'.
4. Select and display only the 'Name', 'Age', and 'FarePerClass' columns.

"""

# Add column 'FarePerClass'
titanic_df["FarePerClass"] = titanic_df["Fare"] / titanic_df["Pclass"]
display(titanic_df["FarePerClass"])

# Filter adult male passengers (Age  18)
adult_males = titanic_df[(titanic_df["Sex"] == "male") & (titanic_df["Age"] >= 18)]
display(adult_males)

# Sort adult males by FarePerClass in descending order
adult_males_srt = adult_males.sort_values(by="FarePerClass", ascending=False)
display(adult_males_srt)

# Select only 'Name', 'Age', and 'FarePerClass'
result = adult_males_srt[["PassengerId","Name", "Sex", "Age", "Ticket", "Cabin","FarePerClass"]]

# Display the result
print("\n\nAdult male passengers with highest fare relative to class:")
display(result)

```

# 3.3 Exploratory Data Analysis with Group-by Method Practice Exercise:

## What percent of the total fare revenue came from each passenger class?
To answer the question perform following operation:
1. Calculate the total Fare paid across all passengers in the Titanic dataset.
2. Subset for passengers in first class (Pclass is 1) and calculate their total fare.
3. Do the same for second class (Pclass is 2) and third class (Pclass is 3).
4. Combine the fare totals from first, second, and third classes into a list.
5. Divide the totals for each class by the overall total fare to get the proportion of fare revenue by class.

```
"""
Problem: Calculate what percent of the total fare revenue came from each passenger class.

Workflow:
1. Compute the total fare collected from all passengers.
2. Calculate the total fare collected from first-class passengers (Pclass = 1).
3. Calculate the total fare collected from second-class passengers (Pclass = 2).
4. Calculate the total fare collected from third-class passengers (Pclass = 3).
5. Combine the totals into a list and compute the proportion of each class relative to the overall total fare.
6. Display the result as percentages.

Author: Bipul Regmi

"""

# Total fare across all passengers
total_fare = titanic_df["Fare"].sum()
print("\n\nTotal fare collected by all passengers:", total_fare)


# Total fare by class
fare_first_class = titanic_df[titanic_df["Pclass"] == 1]["Fare"].sum()
print("\n\nTotal fare collected by first class passengers:", fare_first_class)
fare_second_class = titanic_df[titanic_df["Pclass"] == 2]["Fare"].sum()
print("\n\nTotal fare collected by second class passengers:", fare_second_class)
fare_third_class = titanic_df[titanic_df["Pclass"] == 3]["Fare"].sum()
print("\n\nTotal fare collected by third class passengers:", fare_third_class)

# Combine into a list
fare_by_class = [fare_first_class, fare_second_class, fare_third_class]

# Calculate percentage of total fare for each class
fare_percentage = [(f / total_fare) * 100 for f in fare_by_class]


# Display the results
print("\n\nPercentage of total fare revenue by class:")
print(f"First Class: {fare_percentage[0]:.2f}%")
print(f"Second Class: {fare_percentage[1]:.2f}%")
print(f"Third Class: {fare_percentage[2]:.2f}%")
```

## What percent of the total number of passengers on the Titanic belonged to each age group (e.g., child, adult, senior)?
To answer the question perform following operation:
1. Create a new column, age group, that categorizes passengers into "child" (age < 18), "adult" (age
18{64), and "senior" (age 65 and above).
2. Calculate the total number of passengers on the Titanic.
3. Count the number of passengers in each age group.
4. Divide the count of each age group by the total number of passengers to get the proportion of passengers
in each age group.
5. Display the proportion as a percentage.

```
"""
Problem: Calculate what percent of the total number of passengers on the Titanic belonged to each age group.

Workflow:
1. Create a new column 'AgeGroup' that categorizes passengers as:
   - "child" for age < 18
   - "adult" for age 1864
   - "senior" for age  65
2. Calculate the total number of passengers.
3. Count the number of passengers in each age group.
4. Compute the proportion of each age group relative to the total number of passengers.
5. Display the results as percentages.

Author: Bipul Regmi
"""

# Create age group column
titanic_df["AgeGroup"] = pd.cut(titanic_df["Age"], bins=[0, 17, 64, 200], labels=["child", "adult", "senior"])

# Total passengers
total_passengers = titanic_df.shape[0]

# Count of passengers by age group
age_group_counts = titanic_df["AgeGroup"].value_counts()

# Percentage of passengers in each age group
age_group_percentage = (age_group_counts / total_passengers) * 100

# Display results
print("\n\nPercentage of passengers by age group:")
for group, percent in age_group_percentage.items():
    print(f"{group.capitalize()}: {percent:.2f}%")
```
